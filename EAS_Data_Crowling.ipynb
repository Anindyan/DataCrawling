```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
browser.get("https://bwf.tournamentsoftware.com/ranking/ranking.aspx?rid=70")
soup = BeautifulSoup(browser.page_source,"lxml")
```

```python
rows = soup.find_all('tbody')
rows
```

```python
data = []
for column in rows :
    doubles = column.find(class_='doubles').text
    
    dt = {
        'doubles': doubles,
        'doubles': doubles
    }
    data.append(dt)
```
